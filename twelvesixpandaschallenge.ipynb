```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Create dataset
data = {
    'Product': np.random.choice(['A', 'B', 'C', 'D', 'E'], size=50),
    'Category': np.random.choice(['Electronics', 'Clothing', 'Furniture'], size=50),
    'Sales': np.random.randint(50, 500, size=50),
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], size=50),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], size=50)
}
df = pd.DataFrame(data)
df.to_csv('sales_data.csv', index=False)
df
```

```python
# Apply styling
styled_df = df.style \
    .set_properties(**{
        'font-size': '14pt',          # Font size
        'text-align': 'center',       # Alignment
        'color': 'darkblue',          # Text color
        'background-color': 'lightgray' # Cell background
    })

# Display the styled DataFrame
styled_df

```

```python
july_sales = df[df['Month'] == 'Jul']['Sales'].sum()
july_sales
```

```python
North_sales=df[df['Region']=='North']['Sales'].sum()
North_sales
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_data.csv')
sales_by_product = df.groupby('Month')['Sales'].sum()
sales_by_product.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
df=pd.read_csv('sales_data.csv')
Region_sales=df.groupby('Region')['Sales'].sum()
Region_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('sales in regions')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```
